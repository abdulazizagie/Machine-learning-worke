# missing-category-impputation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('home_saling.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts().sort_values(ascending=False).plot.bar()
plt.xlabel('GaregeQual')
plt.ylabel('number of houses')
```

```python
df['GarageQual'].fillna('Missing',inplace=True)
```

```python
df['GarageQual'].value_counts().sort_values(ascending=False).plot.bar()
plt.xlabel("GaragehQual")
plt.ylabel('number of houses')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['SalePrice']),df['SalePrice'],test_size=0.2,random_state=0)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='constant' , fill_value='Missing')
```

```python
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

```python
imputer.statistics_
```
